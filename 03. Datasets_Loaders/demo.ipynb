```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
```

```python
training_data = datasets.FashionMNIST(
    root="../data",
    download=True,
    train= True,
    transform= ToTensor()
)

test_data = datasets.FashionMNIST(
    root="../data",
    download= True,
    train= False,
    transform= ToTensor()
)
```
